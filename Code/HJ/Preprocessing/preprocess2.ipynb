```python
import pandas as pd
import numpy as np
from tqdm import tqdm, notebook
import time
import warnings

warnings.filterwarnings(action='ignore')
```

```python
# df = 횡령배임 크롤링한 dataframe
df = pd.read_csv(r'C:\apps\Ubion-CorporateBankruptcyAnalysis\Code\HJ\Selenium\excel_data\0020공시67.csv',index_col=0)
df
```

```python
# dataset = 원본 데이터셋
dataset = pd.read_csv(r'C:\apps\Ubion-CorporateBankruptcyAnalysis\Data\최종데이터셋\Finaldataset_ver3.csv',index_col=0)
dataset
```

---

```python
# df에 키값 만들기 : 거래소코드 + 공시발생일
df['key']='0'           # 'key' 컬럼 생성
df
```

```python
for i in range(len(df['거래소코드'])):
    df['key'][i] = str(df['거래소코드'][i])+str(df['공시발생일'][i][0:4])
df
```

```python
# df의 key값이 모두 담긴 리스트 생성
df_key_list=[]

for i in range(len(df['key'])):
    df_key_list.append(df['key'][i])
```

```python
df_key_list
```

```python
# df_key_list 길이 = 830
len(df_key_list)
```

```python
# 원본 데이터셋에 '횡령배임' 컬럼 생성
dataset['횡령배임']=0
dataset
```

```python
# 원본 데이터셋의 'key' 컬럼 -> int형에서 str형으로 형변환
dataset['key'] = dataset['key'].astype('str')
```

```python
type(df['key'][0])
```

## 방법1 : 이중 for문

- df_key_list에 담긴 key값과 원본 데이터셋의 'key' 컬럼의 값 비교해서<br>
    값이 일치하면 '횡령배임' 컬럼의 값을 value 변수에 저장하고 1 증가시킨 후, 다시 '횡령배임' 컬럼에 저장.

```python
for i in tqdm(range(len(df_key_list))):                # 830개
    for j in tqdm(range(len(dataset['key']))):         # 27153개
        if(df_key_list[i]==dataset['key'][j]):
            value=dataset['횡령배임'][j]
            value=value+1
            dataset['횡령배임'][j]=value
```

## 방법2 : value_count()로 딕셔너리 생성 후 merge

```python
# df['key']에 중복 존재

df['key'].value_counts()
```

```python
# df['key'].value_counts()의 인덱스를 key컬럼의 값으로, df['key'].value_counts()의 value를 횡령배임컬럼의 값으로 하는
# 데이터 프레임 생성

subset = pd.DataFrame({'key' : df['key'].value_counts().index, '횡령배임' : df['key'].value_counts().values} )
subset
```

```python
# 원본데이터셋과 subset을 'key'컬럼 기준으로 merge
merged_dataset = dataset.merge(subset, how='outer', on='key')
merged_dataset
```

```python
# dropna를 하기 위해 '횡령배임_y'컬럼 값 0으로 채움.
merged_dataset['횡령배임_y'] = merged_dataset['횡령배임_y'].fillna(0)
```

```python
merged_dataset.dropna(inplace=True)
```

```python
merged_dataset.drop('횡령배임_x', axis=1, inplace=True)
```

```python
# '횡령배임_y'에서 '횡령배임'으로 컬럼 이름 변경
merged_dataset.rename(columns = {'횡령배임_y' : '횡령배임'}, inplace=True)
```

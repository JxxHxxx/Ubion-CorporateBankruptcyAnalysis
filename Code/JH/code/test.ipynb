### 회사명 NaN 제거 

```python
train.dropna(subset=['회사명'], axis = 0, inplace=True)
```

### 부도컬럼 생성

```python
train['부도'] = ""
```

감사의견거절 감사보고서 부적정, 의견거절, 범위제한한정
- 의견코드 : AG, DS, DI, DU, QS

```python
train[['소속코드','감사의견코드','공시제목']]
```

```python
con1 = train['소속코드'] == '4.0'
con2 = train['감사의견코드'] == 'DS'
con3 = train['공시제목'] != 'NaN'


train['부도'] = train.apply[con1 | con2 | con3] = 1
train['부도']
```

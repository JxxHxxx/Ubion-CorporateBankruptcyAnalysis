## **라이브러리**

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\Users\JH\Desktop\CT\0320공시44_전처리완.csv', index_col=0)

dataset['info'] = ''
```

# **함수**

```python
def findword(word, num = 100): # 필요한 텍스트 정보 근방을 추출
    info = []

    for i in range(len(dataset)):
        if dataset['info'][i] == '':
            k = dataset['text'][i].find(word)
            info.append(dataset['text'][i][k:k+num])
        else:
            info.append('')

    for i in range(len(dataset)):
        if dataset['info'][i] == '':
            dataset.at[i,'info'] = info[i]
        else: 
            pass
        
    print('남은 데이터 수 :',dataset['info'].value_counts()[0])
```

```python
word_arr = ['자금조달의목적시설자금','조달된자금의구체적인사용내역','자금의용도','자금조달의목적기타자금','자금조달의']

for word in word_arr:
    findword(word)

dataset.drop(index = dataset.loc[dataset['info'] == ''].index, inplace=True)
dataset.reset_index(inplace=True)
dataset.drop('level_0', axis=1)
```

# **채권 발행 목적을 시설, 운영, 기타 3가지로 구분**

```python
cnt = []
for i in range(len(dataset)):
    idxoper = dataset['info'][i].find('운영자금(원)') + len('운영자금(원)')
    idxist = dataset['info'][i].find('시설자금(원)') + len('시설자금(원)')
    try:
        if dataset['info'][i][idxist].isdigit() == True:
            cnt.append('시설')
        elif dataset['info'][i][idxoper].isdigit() == True:
            cnt.append('운영')
        else:
            cnt.append('기타')
    except:
        cnt.append('없음')

dataset['주식관련사채'] = cnt
```

```python
subset_ = dataset[['key','주식관련사채']]

subset_1 = subset_.loc[subset_['주식관련사채'] == '운영']
subset_2 = subset_.loc[subset_['주식관련사채'] == '기타']
subset_3 = subset_.loc[subset_['주식관련사채'] == '시설']

oper = subset_1['key'].value_counts()
infr = subset_3['key'].value_counts()
etc = subset_2['key'].value_counts()

operdf = pd.DataFrame({'key': oper.index, '신종채권_운영': oper.values})
infrdf = pd.DataFrame({'key': infr.index, '신종채권_시설': infr.values})
etcdf = pd.DataFrame({'key': etc.index, '신종채권_기타': etc.values})

dataset = operdf.merge(infrdf, how='outer', on = 'key')
dataset = dataset.merge(etcdf, how='outer', on = 'key')

dataset.fillna(0, inplace=True)
```

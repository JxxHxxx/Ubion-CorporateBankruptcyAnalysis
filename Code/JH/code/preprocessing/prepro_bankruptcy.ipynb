# **primary key 만들기**

```python
# bankrupt df 의 primary key를 만들기 위한 작업
bankrupt['년'] = bankrupt['시간']
bankrupt = bankrupt.reset_index()

for i in range(len(bankrupt)):
    bankrupt['년'][i] = bankrupt['시간'][i][0:4]

bankrupt['key'] = bankrupt['종목코드'].astype(str) + bankrupt['년'].astype(str)
```

```python
# primary key CASE 1 : 거래소코드(55012) + [회계년도(2012) + 1](2013) ex) 550122013

train['년'] = (train['회계년도'].str[0:4].astype(int) + 1).astype(str)
train_full['년'] = (train_full['회계년도'].str[0:4].astype(int) + 1).astype(str)

# primary key CASE 2 : 거래소코드(55012) + [회계년도(2012)](2012) ex) 550122012

""" train['년'] = (train['회계년도'].str[0:4].astype(int)).astype(str)
train_full['년'] = (train_full['회계년도'].str[0:4].astype(int)).astype(str) """

train['key'] = train['거래소코드'].astype(str) + train['년'].astype(str)
train_full['key'] = train_full['거래소코드'].astype(str) + train_full['년'].astype(str)
```

# **부도 공시 기업 데이터 EDA**

# **train&bankrupt 테이블 조인**

```python
train_bankrupt_merge = pd.merge(train, bankrupt, on='key', how = 'outer')
train_bankrupt_merge
```

```python
train_full_bankrupt_merge = pd.merge(train_full, bankrupt[['key','공시제목']], on='key', how = 'outer')
train_full_bankrupt_merge.head()
```

```python
print(train_full.shape, bankrupt.shape, train_full_bankrupt_merge.shape)
# merge시 outer에 해당하는 부분
train_full.shape[0] + bankrupt.shape[0] - train_full_bankrupt_merge.shape[0]
```

```python
print(train.shape, bankrupt.shape, train_bankrupt_merge.shape)
# merge시 outer에 해당하는 부분
train.shape[0] + bankrupt.shape[0] - train_bankrupt_merge.shape[0]
```

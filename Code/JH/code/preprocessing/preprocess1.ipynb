```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\JH\Desktop\Final\CBA\Data\최종데이터셋\Finaldataset_ver2.csv')
df
```

```python
cp_name=[]
cp_name = df['회사명'].unique().tolist()
cp_name
```

```python
len(cp_name)
```

```python
cnt=[]  # 거래소코드 고유값 개수 담은 리스트 
for name in cp_name:
    cnt.append(len(df.loc[df['회사명']==name]['거래소코드'].unique()))
```

```python
cnt
```

```python
len(cnt)
```

```python
cp_dup_index=[] # 거래소코드 2개이상인 기업의 인덱스 리스트

for i in range(len(cnt)):
    if(cnt[i]>=2):
        cp_dup_index.append(i)
    elif(cnt[i]==0):
        print(0)
    else:
        pass
```

```python
len(cp_dup_index) # 거래소코드 2개이상인 기업 개수
```

```python
cp_dup_index
```

```python
cp_dup_name=[] # 거래소코드 2개이상인 기업명 리스트

for idx in cp_dup_index:
    cp_dup_name.append(cp_name[idx]) 
```

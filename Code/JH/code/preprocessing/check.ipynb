## **사유가 발생했는데 상폐처리안된 케이스**

```python
def get_data(name, code):
    print(dataset.loc[dataset['회사명'] == name]['부도'])
    result = subset.iloc[[code]][['회사명','key','감사의견코드','부도']]
    return result

def check_data(name):
    result = dataset.loc[dataset['회사명'] == name]['부도']
    return result
```

# **RNN 구현하기**

이진 분류의 문제의 경우 출력층의 활성화 함수로 시그모이드 함수를, <br>손실 함수로 binary_crossentropy를 사용합니다. 반면, 다중 클래스 문제라면 출력층의 활성화 함수로 소프트맥스 함수를, 손실 함수로 categorical_crossentropy를 사용합니다. <br>또한, 다중 클래스 분류 문제의 경우에는 클래스가 N개라면 출력층에 해당되는 밀집층(dense layer)의 크기는 N으로 합니다. 즉, 출력층의 뉴런의 수는 N개입니다.

```python
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense
from tensorflow.keras.models import Sequential
```

```python
dataset.head(3)
```

```python
dataset.shape
```

```python
embedding_dim = 5
data_size = 10164
```

```python
model = Sequential()
model.add(Embedding(data_size, embedding_dim))
model.add(SimpleRNN(32)) # RNN 셀의 hidden_size는 32
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
#특수문자 제거 함수 
def elimination_spec (dataset) :
    col_list = list(dataset.columns)

    import re
    for i in range(len(col_list)):
        filtered_col = re.sub(r'[^\d\w]', '_',col_list[i])
        if col_list[i][0] == '1':
            col_list[i] = '_'+ col_list[i]
        else :
            pass
        col_list[i] = filtered_col

    dataset.columns = col_list
    
```

```python
elimination_spec(df)
```

# **한국은행 모형 변수**

- 자기자본배당률, 적립금비율, 금융비용대 부채비율, 자기자본비율, 유동부채비율, 매입채무회전율, 영업자산회전율, 총자본투자효율


```python
#한국은행
'자기자본배당률'
#적립금비율 우리는 적립금을 다루지않아 드랍
'금융비용 대 부채비율' 이자비용/부채
'자기자본배율' 자기자본/총자본
'유동부채비율'  유동부채/자기자본
'매입채무회전률' 매출액/매입채무
'영업자산회전률' #우리는 유형자산회전률 사용
'총자본투자효율' 부가가치/총자본
```

```python
df['금융비용 대 부채비율'] = df['이자비용_요약_']/df['부채총계_요약_']
```

```python
# 운전 자본은 모든 유동 자산의 합계이며 투자자에게 그다지 중요하지 않음. 
# 순 운전 자본은 회사의 유동 부채에 대한 유동 자산의 초과분으로 회사의 재무 건전성에 대한 중요한 지표.
```

```python
#알트만은 교수님께서 강조하신 부도예측의 basic임으로 없는 feature 만들어보기로 함
```

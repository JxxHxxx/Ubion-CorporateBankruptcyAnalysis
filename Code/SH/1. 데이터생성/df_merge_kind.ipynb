## 거래소코드 앞에 붙은 0 제거

```python
# 거래소코드 앞에 붙은 0 제거

def elimination_0(data):
    for i in range(len(data)):
        if data['거래소코드'][i][:1] == '0':
            data['거래소코드'][i] = data['거래소코드'][i].lstrip("0")
    else : pass
```

## key 생성

```python
# key 생성

def key_maker(data):
    data['년'] = data['공시발생일'].str[0:4].astype(int)
    data['key'] = data['거래소코드'].astype(str) + data['년'].astype(str)
```

```python
key_maker(t_11)
key_maker(t_56)
key_maker(t_65)
key_maker(t_67)
```

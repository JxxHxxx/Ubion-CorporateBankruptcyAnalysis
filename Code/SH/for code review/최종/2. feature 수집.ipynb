## **1. 안정성**
A6 부채구성비율 : (부채총계/총자산)*100 <BR>
A11 청산가치율 : (순자산(총자본)/매출액)*100 <BR>
A14 이익잉여금비율 : (이익잉여금/자기자본)*100 <BR>
A15 자본잉여금비율 : (자본잉여금/자기자본)*100 <BR>
A17 경영자산비율 : [(유동자산+유무형자산)/(총자산)]*100 <BR>

-> 부채구성비율, 청산가치율, 이익잉여금비율, 자본잉여금비율, 경영자산비율(5개)

```python
df['부채구성비율'] = (df['부채총계_요약_']/df['자산총계_요약_'])*100
df['청산가치율'] = (df['자본총계_요약_']/df['매출액_요약_'])*100
df['이익잉여금비율'] = (df['이익잉여금_요약_']/df['자본총계_요약_'])*100
df['자본잉여금비율'] = (df['자본잉여금_요약_']/df['자본총계_요약_'])*100
df['경영자산비율'] = ((df['유동자산_요약_']+df['유형자산_요약_']+df['무형자산_요약_'])/df['자산총계_요약_'])*100
```

# **결측치 및 inf(무한대) 값 처리**

### inf 값 처리

```python
df = df.replace([np.inf, -np.inf], np.nan) # replace 메서드로 np.inf를 None(np.nan)으로 변경 print(df)
```

### 결측치 처리

```python
df.isnull().sum()
```

```python
cpylist = df['회사명'].unique()

for i in tqdm(range(len(cpylist))):
    df.loc[df['회사명'] == cpylist[i]] = df.loc[df['회사명'] == cpylist[i]].fillna(method='bfill') #직후 데이터로 대체
    df.loc[df['회사명'] == cpylist[i]] = df.loc[df['회사명'] == cpylist[i]].fillna(method='pad') #직전 데이터로 대체
    df.loc[df['회사명'] == cpylist[i]] = df.loc[df['회사명'] == cpylist[i]].fillna(0) # 그래도 처리못하는 데이터는 0으로 대체
```

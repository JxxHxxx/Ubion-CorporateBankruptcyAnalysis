## Lasso 진행 및 데이터프레임 생성

```python
X = subset.drop(columns=['부도'])
y = subset['부도']
```

```python
from sklearn import linear_model
from sklearn.linear_model import Lasso
import numpy as np
lasso = Lasso().fit(X, y)
print('{:.3f}'.format(lasso.score(X, y)))
```

```python
# 사용한 변수 개수
print('전체 변수 개수 : ' ,np.sum(lasso.coef_.shape))
print('**사용된 변수 개수 : ' ,np.sum(lasso.coef_ != 0))
```

```python
print(lasso.coef_)
```

```python
lasso_coef =lasso.coef_.tolist() # lasso 변수별 계수 리스트
lasso_feature = lasso.feature_names_in_.tolist() # 변수 리스트 
```

```python
print(len(lasso_coef))
print(len(lasso_feature))
```

```python
df_lasso = pd.DataFrame([lasso_feature, lasso_coef])
df_lasso = df_lasso.rename(columns=df_lasso.iloc[0]).drop(df_lasso.index[0]).rename(index={1: 'lasso coef'})
df_lasso
```

```python
df_lasso != 0
```

```python
lasso_n_0 = df_lasso != 0
lasso_n_0=lasso_n_0.rename(index={'lasso coef' : 'used'})
```

```python
lasso_n_0
```

```python
df_lasso = pd.concat([df_lasso, lasso_n_0]).T
```

```python
df_lasso
```

```python
df_lasso_final = df_lasso[df_lasso['used'] == True]
```

```python
# 논문에서 lasso 계수 절대값이 높으면 영향력이 높음 
df_lasso_final['abs_coef'] = abs(df_lasso_final['lasso coef'])
```

## 랭킹 칼럼 생성

```python
# 논문에서 lasso 계수 절대값이 높으면 영향력이 높다고 언급해서 abs로 진행 
df_lasso_final['abs_coef'] = abs(df_lasso_final['lasso coef'])
df_lasso_final['rank'] = df_lasso_final['abs_coef'].rank(method='max').astype(int)
df_lasso_final
```

```python
lasso_rank = df_lasso_final.sort_values('rank')
lasso_rank
```

```python
print('사용된 변수의 개수 : ',len(lasso_rank))
print('사용된 변수 : ', lasso_rank.index)
```

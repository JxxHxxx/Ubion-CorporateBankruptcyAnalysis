# **회계년도가 짧은 회사 모음**

```python
cpyname = dataset.loc[dataset['부도'] == 1]['회사명'].unique()

shtcpy = [] # 회계기간이 짧은 회사 list
shtidx = [] # 회계기간이 짧은 회사의 idx
for i in cpyname:
    if len(dataset.loc[dataset['회사명'] == i]) < 5:
        shtcpy.append(i)
for i in shtcpy:
    shtidx.append(dataset.loc[dataset['회사명'] == i].index[-1])

shtidx
```

# **N년 전 부도 데이터 만들기**

```python
def makeNyear(num):
    cpyidx = []
    bfidx = []
    for idx_ in dataset.loc[dataset['부도'] == 1].index:
        cpyidx.append(idx_)

    for name in shtidx:
        try:
            cpyidx.remove(name)
        except ValueError:
            print(name)
            

    for idx_ in cpyidx:
        bfidx.append(idx_ - num)

    dataset.at[cpyidx, '부도'] = 0
    dataset.at[bfidx, '부도'] = 1

    dataset.to_csv(r'C:\Users\JH\Desktop\Final\Ubion-CorporateBankruptcyAnalysis\Code\JH\data\dataset_{num}year_ago.csv'.format(num = num))
```

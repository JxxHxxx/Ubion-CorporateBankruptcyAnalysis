### **회사명 NaN 제거**

```python
train.dropna(subset=['회사명'], axis = 0, inplace=True)
```

### **상장폐지 중 부도기업 추출**

감사의견거절 감사보고서 부적정, 의견거절, 범위제한한정
- 의견코드 : AG, DS, DI, DU, QS

```python
df = train[['소속코드','감사의견코드','공시제목','회사명']]
df = df.fillna(0)
```

```python
con1 = df['소속코드'] == 4
con2 = (df['감사의견코드'] == 'DS')|(df['감사의견코드'] =='DI')|(df['감사의견코드'] =='DU')|(df['감사의견코드'] =='AG')|(df['감사의견코드'] =='QS')
con3 = df['공시제목'] != 0
```

```python
df_bankrupt = df.loc[con1&(con2|con3)]
```

```python
df_bankrupt
```

### **부도컬럼 생성**

```python
train['부도'] =""
```

```python
df_bankrupt
```

```python
train['부도'] = train.loc[con1&(con2|con3), '부도'] == 1
```

```python
train['부도'].value_counts()
```

```python
train['부도'] = train['부도'].fillna(1)
train['부도'] = train['부도'].replace(False, 0)
train['부도'].value_counts()
```

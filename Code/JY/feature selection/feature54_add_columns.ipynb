# **key 값 생성**

```python
# key 생성
def key_maker(data):
    data['년'] = data['회계년도'].str[0:4].astype(int)
    data['key'] = data['거래소코드'].astype(str) + (data['년']+1).astype(str)
```

```python
key_maker(df)
```

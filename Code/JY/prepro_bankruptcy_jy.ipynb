# **primary key 만들기**
- 상장폐지 사유 발생 및 사유발생과의 시점차이가 존재 하므로 전처리 편의상 '년'변수 생성시 +1 

#### train, train_full : 회계년도 + 1 = '년' 변수 추가

```python
train['년'] = (train['회계년도'].str[0:4].astype(int) + 1).astype(str)
train_full['년'] = (train_full['회계년도'].str[0:4].astype(int) + 1).astype(str)
```

```python
# 확인
train.head()
```

#### train, train_full : 'key' 변수 추가 

```python
train['key'] = train['거래소코드'].astype(str) + train['년'].astype(str)
train_full['key'] = train_full['거래소코드'].astype(str) + train_full['년'].astype(str)
```

```python
# 확인
train.head()
```

#### bankrupt : '년' & 'key' 변수 추가

```python
bankrupt.info()
```

```python
bankrupt['년'] = bankrupt['시간']
bankrupt = bankrupt.reset_index()
```

```python
for i in range(len(bankrupt)):
    bankrupt['년'][i] = bankrupt['시간'][i][0:4]
```

```python
bankrupt['key'] = bankrupt['종목코드'].astype(str) + bankrupt['년'].astype(str)
```

```python
#확인
bankrupt
```

```python
tt = pd.merge(train, bankrupt, on='key', how = 'outer')
tt
```
